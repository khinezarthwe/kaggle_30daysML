```python
numeric_train = train_data.select_dtypes(include=["float64", "int64"])
numeric_test = test_data.select_dtypes(include=["float64", "int64"])
categorical_train=train_data.select_dtypes(include='object')
categorical_test=test_data.select_dtypes(include='object')
```

```python
object_cols = [col for col in train_data.columns if train_data[col].dtype == "object"]

object_nunique = list(map(lambda col: categorical_train[col].nunique(), object_cols))
d = dict(zip(object_cols, object_nunique))

# Print number of unique entries by column, in ascending order
sorted(d.items(), key=lambda x: x[1])
```
